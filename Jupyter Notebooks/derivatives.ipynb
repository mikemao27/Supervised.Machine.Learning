```python
from sympy import symbols, diff
```

```python
J = (3) ** 2
J_epsilon = (3 + 0.001) ** 2
k = (J_epsilon - J) / 0.001
print(f"J = {J}, J_epsilon = {J_epsilon}, dJ_dw ~= k = {k:0.6f} ")
```

```python
J = (3) ** 2
J_epsilon = (3 + 0.000000001) ** 2
k = (J_epsilon - J) / 0.000000001
print(f"J = {J}, J_epsilon = {J_epsilon}, dJ_dw ~= k = {k}")
```

```python
J, w = symbols("J, w")
```

```python
J = w ** 2
J
```

```python
dJ_dw = diff(J, w)
dJ_dw
```

```python
dJ_dw.subs([(w, 2)])
```

```python
dJ_dw.subs([(w, 3)])
```

```python
dJ_dw.subs([(w, -3)])
```
